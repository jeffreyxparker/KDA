```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
responses = pd.read_csv("survey_responses_1st_survey.csv")
responses = responses.drop_duplicates(subset = "id", keep = "first")
```

```python
responses.head()
```

```python
seg_data = responses.loc[:,'Delicious vs Healthy':'Deal Driven vs Regular Menu']
seg_data = pd.get_dummies(seg_data)
seg_data = seg_data.iloc[:,[0,3,5,6]]
seg_data.columns = ['Delicious','Traditional','Shop_Loyal','Deal_Driven']
```

```python
seg_data
```

```python
from sklearn.cluster import KMeans

k3 = KMeans(n_clusters=3)
k3.fit(seg_data)
```

```python
print(k3.labels_)
```

```python
seg_data['cluster'] = k3.labels_
```

```python
seg_data
```

```python
seg_data.groupby(["cluster","Delicious"]).size()
```
